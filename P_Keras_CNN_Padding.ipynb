```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense, Conv2D,Flatten
from keras import Sequential
from keras.datasets import mnist
```

```python
(x_train,y_train), (x_tes,y_test)=mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding="valid",activation="relu", input_shape=(28,28,1)))# 32 are filtter ,
#kernal_size = filter size,valid=no padding, 1 = black & white image
model.add(Conv2D(32,kernel_size=(3,3),padding="valid",activation="relu"))
model.add(Conv2D(32,kernel_size=(3,3),padding="valid",activation="relu"))

model.add(Flatten())

model.add(Dense(128,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```python
model.summary()
```

```python
model =Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding="same",strides=(2,2),activation="relu",input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding="same",strides=(2,2),activation="relu"))
model.add(Conv2D(32,kernel_size=(3,3),padding="same",strides=(2,2),activation="relu"))

model.add(Flatten())

model.add(Dense(128,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```python
model.summary()
```
